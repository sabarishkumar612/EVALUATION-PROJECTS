```python
#IMPORTING THE REQUIRED LIBRARIES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
ds3 = pd.read_csv('countrycode.csv')
ds3
```

```python
ds2 = pd.read_csv('zomatofile.csv')
ds2
```

```python
ds1 = pd.merge(ds2,ds3, on="Country Code")
ds1
```

```python
ds1=ds1.drop(columns=['Country Code','Address','Locality','Longitude','Latitude','Restaurant Name','City','Locality Verbose','Is delivering now','Switch to order menu','Rating color'],axis=1)
ds1
```

```python
ds1.set_index("Restaurant ID", inplace = True)
```

```python
#CHECKING FOR NULL VALUES
ds1.isnull().sum()
```

```python
print(ds1['Cuisines'].mode())
```

```python
#HANDLING NULL VALUES
#Applying Mode imputation technique 
ds1['Cuisines'].fillna(ds1['Cuisines'].mode()[0],inplace=True)
ds1
```

```python
#CHECKING FOR NULL VALUES
ds1.isnull().sum()
```

```python
ds1.info()
```

```python
#CHECKING FOR THE COUNT OF EVERY VALUES IN EACH COULMNS
for i in ds1.columns:
    print(ds1[i].value_counts())
```

```python
ds1.nunique()
```

```python
ds1.describe()
```

- There are no null values in the dataset
- Huge difference mean and median indicates presence of skewness
- Huge difference between 75% and max values and hence outliers are present

```python
#SEPERATING CATEGORICAL AND NUMERICAL COLUMNS
cat_col=[]
num_col=[]
for i in ds1.dtypes.index:
    if ds1.dtypes[i]=="object":
        cat_col.append(i)
    else:
        num_col.append(i)
print(cat_col)
print(num_col)
```

```python
#CHECKING FOR OUTLIERS
from scipy.stats import zscore
f=ds1[['Votes','Aggregate rating','Price range','Average Cost for two']]
z=np.abs(zscore(f))
z
```

```python
print(np.where(z>3))
```

```python
#REMOVING OUTLIERS
ds6=ds1[(z<3).all(axis=1)]
ds6
```

```python
ds6.skew()
```

```python
#REMOVING SKEWNESS USING CUBEROOT METHOD
ds6[['Votes']] = np.cbrt(ds6[['Votes']])
ds6.skew()
```

```python
# LABEL ENCODING TECHNIQUE
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in cat_col:
    ds6[i]=le.fit_transform(ds6[i].astype(str))
ds6
```

```python
ds6.corr()
```

```python
#SEPERATING INDEPENDENT VARIABLES AND TARGET
x1 = ds6.drop('Average Cost for two',axis=1)
y1 = ds6['Average Cost for two']
print(x1)
```

```python
y1
```

```python
#SCALING TO REMOVE BIASNESS - STANDARD SCALING
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
x1=st.fit_transform(x1)
x1
```

```python
x1=pd.DataFrame(data=x1)
x1
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(x1.values, i) for i in range(x1.shape[1])]
vif["features"] = x1.columns
vif
```

# MODELLING - AVERAGE COST FOR TWO AS TARGET

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.40,random_state=42)
```

```python
#LINEAR REGRESSION
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score,KFold
m1=LinearRegression()
m1.fit(x1_train,y1_train)
trainprediction1=m1.predict(x1_train)
print(r2_score(y1_train,trainprediction1))
testprediction1=m1.predict(x1_test)
print(r2_score(y1_test,testprediction1))
kfold=KFold(n_splits=5,shuffle=True)
score1=cross_val_score(m1,x1,y1,cv=kfold)
print(score1)
print(score1.mean())
print((r2_score(y1_test,testprediction1)-(score1.mean())))
```

```python
#ADA BOOST REGRESSION
from sklearn.ensemble import AdaBoostRegressor
m2=AdaBoostRegressor()
m2.fit(x1_train,y1_train)
trainprediction2=m2.predict(x1_train)
print(r2_score(y1_train,trainprediction2))
testprediction2=m2.predict(x1_test)
print(r2_score(y1_test,testprediction2))
kfold=KFold(n_splits=5,shuffle=True)
score2=cross_val_score(m2,x1,y1,cv=kfold)
print(score2)
print(score2.mean())
print((r2_score(y1_test,testprediction2)-(score2.mean())))
```
